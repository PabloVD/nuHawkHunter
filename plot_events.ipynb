# PBH event rates

Plot the event rate of neutrinos for different neutrino detectors

```python
from Source.event_rate import *

Mpbhs =  [1e15, 2e15, 4e15]

fpbhs = 1.e-2*np.ones_like(Mpbhs)
cols = ["r", "purple", "b", "g", "orange"]

list_exps = ["HK", "JUNO", "DUNE"]

def plot_events(Mpbhs, fpbhs, exp):

    compute_events(Mpbhs, fpbhs, exp, plotevents=1)

    plt.ylim(1.e-3, 2.e2)
    plt.yscale("log")
    plt.legend(loc="lower right")
    plt.xlabel('$E{\\rm \,\, [MeV]}$')
    plt.ylabel('${\\rm d}N/d E \,\, [{\\rm MeV}^{-1}{\\rm yr}^{-1}]$')
    plt.title(exp)
    plt.savefig("figures/events_"+exp+".png", bbox_inches='tight')
    plt.show()
    plt.gcf().clear()

for exp in list_exps:
    plot_events(Mpbhs, fpbhs, exp)
```
